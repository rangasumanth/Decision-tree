```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_excel("Baseball_salary.xlsx")
data
```

```python
data.rename(columns={'Unnamed: 0': 'Names'},inplace=True)
data
```

```python
data['Names'] = data['Names'].apply(lambda x: x.strip('-'))
data
```

```python
#Renaming the header
data.rename(columns={'CAtBat':'CareerAtBat','CHits':'CareerHits','CHmRun':'CareerHR','CRuns':'CareerRuns','CRBI':'CareerRBI','CWalks':'CareerWalks'},inplace=True)
data
```

```python
data = data.dropna()
```

```python
data
```

```python
sns.histplot(data.Salary)
```

```python
# Assuming your DataFrame is named 'data'
numerical_data = data.select_dtypes(include=['int64', 'float64'])

# Calculate the correlation matrix
correlation_matrix = numerical_data.corr()

# Plotting the correlation matrix as a heatmap
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Matrix")
plt.show()
```

```python
# scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(data['Hits'], data['Salary'], alpha=0.5)
plt.xlabel('Hits')
plt.ylabel('Salary')
plt.title('Scatter Plot between Feature1 and Salary')
plt.show()
```

```python
# scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(data['Hits'], data['Runs'], alpha=0.5)
plt.xlabel('Hits')
plt.ylabel('Salary')
plt.title('Scatter Plot between Feature1 and Salary')
plt.show()
```
